```python
class Node(object):
    def __init__(self):
        self.sets = list()
        self.solus = list()
        self.commen = None
        
    def __str__(self):
        return str(int(bool(self.sets)))
    
    def __repr__(self):
        return str(int(bool(self.sets)))
    
    def __bool__(self):
        return bool(self.sets)
    
    def add(self, _set, _solution):
        """
        _set must be a number a hash value
        """
        if _set not in self.sets:
            self.sets.append(_set)
            self.solus.append(_solution)
            
    def merge(tar, x, y, w1, h1, w2, h2, itn):
        for ix, x_set in enumerate(x.sets):
            for iy, y_set in enumerate(y.sets):
                s = [i + j for i, j in zip(int2set[x_set], int2set[y_set])]
                flag = True
                for it, w in enumerate(s):
                    if w > upper_bound[it]:
                        flag = False
                        break
                if flag:
                    _set = set2int(str(s))
                    tar.add(_set, (w1, h1, w2, h2, ix, iy, itn))
                    int2set[_set] = s
    def show(self):
        for _set, _solus in zip(self.sets, self.solus):
            print (int2set[_set], _solus, sep="\n")
```

```python
W = 60
H = 90
N = sum(upper_bound)
```

```python
dp = [[[Node() for k in range(N+1)] for i in range(H+1)] for j in range(W+1)]
```

```python
for it, (w, h, n) in enumerate(blocks):
    sets = [0]*len(blocks)
    sets[it] = 1
    _set = set2int(str(sets))
    int2set[_set] = sets
    if w <= W and h <= H:
        dp[w][h][1].add(_set, tuple())
    if h <= W and w <= H:
        dp[h][w][1].add(_set, tuple())
```

```python
NUM = 2
# NUM should be a function output with the scale of area and n
for i in range(1, W+1):
    for j in range(1, H+1):
        for n in range(1, N+1):
            for it in range(1, i//2+1):
                for itn in range(1, n):
                    merge(dp[i][j][n], dp[it][j][itn], dp[i-it][j][n-itn], it, j, i-it, j, itn)
            for jt in range(1, j//2+1):
                for jtn in range(1, n):
                    merge(dp[i][j][n], dp[i][jt][jtn], dp[i][j-jt][n-jtn], i, jt, i, j-jt, jtn)
            
            dp[i][j][n].sets = random.sample(dp[i][j][n].sets, min(len(dp[i][j][n].sets), NUM))
            now_percent = (i-1)/W + (j-1)/(H*W) + (n-1)/(N*W*H)
            nn = int(now_percent*50)
            string = '=' * nn
            e_string = '.' * (50 - nn)
            print ("\r|" + string + '>>{0:4.2f}%'.format(now_percent*100)+e_string+"|", end="", flush=True)
```
